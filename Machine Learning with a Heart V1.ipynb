```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import log_loss
from sklearn.model_selection import train_test_split
import time
```

```python
train_values = pd.read_csv("train_values.csv")
```

```python
train_labels = pd.read_csv("train_labels.csv")
```

```python
test_values = pd.read_csv("test_values.csv")
```

```python
x = train_values.drop(columns="patient_id")
```

```python
x.head()
```

```python
train_values.shape
```

```python
x.shape
```

```python
x['thal']=x['thal'].map({'normal':1,'reversible_defect':2, 'fixed_defect':3 })
```

```python
x.head()
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
parameter=[{'max_features':[2,3,4],
            'min_samples_leaf':[1,2]}]

RF= RandomForestClassifier(n_jobs=-1,n_estimators=200, random_state=1024,oob_score=False)

grid=GridSearchCV(RF, parameter, cv=10)

grid.fit(xtrain, ytrain.ravel())

print("Best score of Ridge: ",grid.best_score_ )
```

```python
grid.best_estimator_
```

```python
test = test_values
```

```python
id_index = test["patient_id"]
test['thal']=test['thal'].map({'normal':1,'reversible_defect':2, 'fixed_defect':3 })
test.drop(['patient_id'], axis=1, inplace=True)
X=np.asarray(test)
```

```python
predictions= grid.predict_proba(X)
pred= predictions[:,1]
```

```python
submission= pd.DataFrame({'patient_id':id_index, 'heart_disease_present':pred})
submission.to_csv("submission.csv", index=False)
```
